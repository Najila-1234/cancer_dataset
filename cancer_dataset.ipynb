```python
import pandas as pd
df=pd.read_csv('/content/cancer.csv')
df
```

```python
df.columns
```

```python
df=df.drop(['compactness_mean', 'concavity_mean','concave points_mean',
        'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst'],axis=1)
df
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,1:].values
x
```

```python
y=df.iloc[:,0].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
sc.fit(x_train)
x_train_new=sc.transform(x_train)
x_train_new
```

```python
x_test_new=sc.transform(x_test)
x_test_new
```

```python
#do knn algorithm
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_new,y_train)
y_pred=knn.predict(x_test_new)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
#naive bayes algorithm
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train_new,y_train)
y_pred_nb=nb.predict(x_test_new)
y_pred_nb
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred_nb))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_nb))
```

```python
#svm algorithm
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train_new,y_train)
y_pred_sv=sv.predict(x_test_new)
y_pred_sv
```

```python
# if a new patient comes predict whether he has cancer or not
sv.predict(sc.transform([[18.5,19.8,130.1,516,.05]]))
```

```python
print(classification_report(y_test,y_pred_sv))
```

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier()
dec.fit(x_train_new,y_train)
y_pred_dec=dec.predict(x_test_new)
y_pred_dec
```

```python
print(classification_report(y_test,y_pred_dec))
```
